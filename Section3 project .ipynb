```python
def kvalue(string):     
   if ',' in string:
      string1 = string.replace(',','')
      return float(string1)
   elif 'N/A' in string:
      string2 = string.replace('N/A','0')
      return float(string2)
   else:
      return float(string)

df['현재가'] = df['현재가'].apply(kvalue)
df['현재가']
```
